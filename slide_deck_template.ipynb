## Investigation Overview

In this investigation of the Prosper Loan Data, I will investigate the relationship between Loan Amount, Loan Terms, and Interest Rates

## Dataset Overview

This data set contains 113,937 loans with 81 variables on each loan, including loan amount, borrower rate (or interest rate), current loan status, borrower income, and many others.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('prosperLoanData_edit.csv')
```

## Investigating Distribution of Loan Amounts

The distribution of loan amount is right-skewed, with a majority of borrowers borrowing under $10,000.

```python
plt.hist(data = df, x = 'LoanOriginalAmount')
plt.title('Ditribution of Loan Amount')
plt.ylabel('Count')
plt.xlabel('Loan Amount ($)');
```

## Investigating the Relationship between Loan Amount and Interest Rate

Now let's take a look at the relationship between the other 2 variables with a scatter plot. Notice that Loan amounts may seem categorical with significantly more values plotted at each of the milestone amounts (5,000, 10,000, 15,000 and so on). It's important to remember that the loan amount values are still numerical and only looks this way due to loans being offered at nice numbers (ending in 0s).

We can clearly see the moderate negative correlation Loan Amount has with the interest rate. While smaller loans (< 10000) are capping out at around 35% interest rate, borrowing 35k is only capping at around 15%. In between these loan amounts, you can see this downward trend very clearly. In fact, the highest amount of interest for someone borrowing 35k is mere 18.19%.

```python
plt.scatter(data = df, x = 'LoanOriginalAmount', y = 'BorrowerRate', alpha = 0.01)
plt.title('Loan Amount vs. Interest Rate')
plt.xlabel('Loan Amount ($)')
plt.ylabel('Borrower Interest Rate');
```

## Investigating the Relationship between Credit Score, Loan Amount, and Interest Rate

A few very interesting insights can be gained here. First, 12 month term loans have the lowest interest rate and is by far the least occuring loan term. The highest recorded 12 month term loan was also 25k. Secondly, we see that based on the opacity of the general graph, 36 term loans are the most frequent - and it only makes sense that the loaners try to make their most money from this group of borrowers by charging them the highest interest across all loans. Across all loan amounts up to 30k, we see that 36 month term loans have the highest interest rates. In the last graph, we see that 60 month loans are the second-most popular loan type. They generally offer a lower interest across the board, except bigger loans, starting at 30k and up, with marginal increase in interest rates.

```python
g = sb.FacetGrid(data = df, col = 'Term', height = 5, col_wrap = 3)
g.map(plt.scatter, 'LoanOriginalAmount', 'BorrowerRate', alpha = 0.05)
g.set_axis_labels("Loan Amount ($)", "Interest Rate")
plt.subplots_adjust(top = 0.85)
g.fig.suptitle('Relationship between Loan Amount & Interest Rate for different Loan Terms');
```
